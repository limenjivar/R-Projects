I had trouble converting LandSlope into a factor, therefore I will just exclude that column. We still have missing values for LotFrontage so we will try to fill the missing values using a decision tree.

```r
#missing values for LotFrontage and removing outliers
boxplot(prices$LotFrontage)
```

We seem to have some outliers so first we will try to get rid of them. 

```r
boxplot.stats(prices$LotFrontage)
```

```r
upper.bound <- boxplot.stats(prices$LotFrontage)$stats[5]
lower.bound <- boxplot.stats(prices$LotFrontage)$stats[1]
lower.clean <- prices$LotFrontage >lower.bound
upper.clean <- lower.clean < upper.bound
```

```r
prices[upper.clean,]
```

```r
library(rpart)
lot.equation <- 'LotFrontage ~ LotArea + Street  + LotConfig + Neighborhood +
                  BldgType + HouseStyle + GrLivArea + GarageYrBlt + GarageArea'
set.seed(123)
lot.model <- rpart(formula = lot.equation, data = prices[upper.clean,], method='class',
                   control = rpart.control(minsplit = 2))
lot.row <- prices[is.na(prices$LotFrontage), 
                  c('LotArea', 'Street', 'LotConfig', 'Neighborhood',
                    'BldgType', 'HouseStyle', 'GrLivArea', 'GarageYrBlt', 'GarageArea')]
lot.pred <- predict(lot.model, newdata = lot.row, type='class') 
prices[is.na(prices$LotFrontage), 'LotFrontage'] <- lot.pred
```

###  Predicting SalePrice

We must first seperate our data back into the train and test set.

```r
#Seperating back into test and training sets
new_train <- prices[prices$IsTrainSet == 1,]
new_test <- prices[prices$IsTrainSet == 0,]
```

```r
library(randomForest)
set.seed(123)
sale.equation <- 'SalePrice ~ MSSubClass  + MSZoning + LotFrontage + LotArea + LotShape + LandContour + Utilities +
Neighborhood + Condition1 + Condition2+BldgType + HouseStyle + OverallQual + OverallCond +YearBuilt +
YearRemodAdd + RoofStyle + RoofMatl + Exterior1st +Exterior2nd +MasVnrType + MasVnrArea + ExterQual +ExterCond +
Foundation + BsmtQual + BsmtCond + BsmtExposure + BsmtFinType1 +BsmtFinSF1 + BsmtFinType2 + BsmtFinSF2 + BsmtUnfSF +
TotalBsmtSF + Heating + HeatingQC + CentralAir + Electrical + X1stFlrSF + X2ndFlrSF + LowQualFinSF + GrLivArea +
BsmtFullBath +  BsmtHalfBath +  FullBath + HalfBath + BedroomAbvGr + KitchenAbvGr + KitchenQual + TotRmsAbvGrd + 
Functional + Fireplaces + FireplaceQu + GarageType  +  GarageYrBlt + GarageFinish + GarageCars + GarageArea + GarageQual + 
GarageCond + PavedDrive + WoodDeckSF + OpenPorchSF +  EnclosedPorch + X3SsnPorch + ScreenPorch +  PoolArea +    
PoolQC + Fence + MiscFeature + MiscVal + MoSold + YrSold + SaleType + SaleCondition'
sale.formula <- as.formula(sale.equation)
regressor <- randomForest(formula = sale.formula, data = new_train, ntree = 500) 

SalePrice <- predict(regressor, newdata = new_test)
```
